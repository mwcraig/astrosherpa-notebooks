```python
import numpy as np
from astropy.modeling.models import Gaussian1D
from itertools import cycle
from astropy.modeling.astro_sherpa import SherpaFitter
from sherpa.stats import Chi2
from collections import  defaultdict
%pylab inline
from itertools import cycle
```

```python
model = Gaussian1D(amplitude=1.2, mean=0.9, stddev=0.5)
err = 0.2
x = np.arange(-3, 3, .08)
y = []
yerrs=[]

y.append(model(x)+err*np.random.uniform(-1,1,size=len(x)))
yerrs.append(err * np.random.uniform(0.5,1,size=len(x)))

model.amplitude = 0.9
model.mean = -0.4
y.append(model(x)+err*np.random.uniform(-1,1,size=len(x)))
yerrs.append(err * np.random.uniform(0.5,1,size=len(x)))


model.amplitude = 2
model.mean = 0
model.stddev = 0.2
```

```python
sfit=SherpaFitter(statistic=Chi2())
xx = np.vstack([x for _ in xrange(len(y))])
fitmodels = sfit(model, xx, y,yerr=yerrs)
```

```python
plt.rcParams['figure.figsize']=(15,10)
plt.plot(x, model(x), lw=2, label="Old Model", c="k")
colcycle = cycle(['r', 'b', 'g', 'c', 'm'])
for yy,yee,mod in zip(y,yerrs,fitmodels):
    plt.errorbar(x, yy, label="Data", c=colcycle.next(),yerr=yee,ls="")
    plt.plot(x,mod(x),label="Fitted model",lw=2)
plt.legend()

```

```python
err_res=sfit.est_errors(sigma=3)

err_vals=defaultdict(list)
for pname, minmax in zip([pname.split(".")[-1] for pname in err_res.parnames], zip(err_res.parmins,err_res.parmaxes)):
    err_vals[pname].append(minmax)

mod_lower_0=fitmodels[0].copy()
mod_upper_0=fitmodels[0].copy()
for pname in err_vals.keys():
    getattr(mod_lower_0,pname).value+=err_vals[pname][0][0]
    getattr(mod_upper_0,pname).value+=err_vals[pname][0][1]


mod_lower_1=fitmodels[1].copy()
mod_upper_1=fitmodels[1].copy()
for pname in err_vals.keys():
    getattr(mod_lower_1,pname).value+=err_vals[pname][1][0]
    getattr(mod_upper_1,pname).value+=err_vals[pname][1][1]
    
```

```python
plt.rcParams['figure.figsize']=(15,10)
plt.ion()
plt.plot(x, model(x), lw=2, label="Old Model", c="k")
colcycle = cycle(['r', 'b', 'g', 'c', 'm'])
for yy,yee,mod in zip(y,yerrs,fitmodels):
    plt.errorbar(x, yy, label="Data", c=colcycle.next(),yerr=yee,ls="",lw=2)
    plt.plot(x,mod(x),label="Fitted model",lw=2)
plt.legend()

for err in [mod_lower_0,mod_upper_0,mod_lower_1,mod_upper_1]:
    plt.plot(x,err(x),label="Error",lw=2,ls="--")
```
